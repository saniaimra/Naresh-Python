```python
import matplotlib.pyplot as plt 
import numpy as np
import math
x = np.linspace(0,2*math.pi,1000)
y = np.sin(x)
y1 = np.sin(2*x)
plt.plot(x,y)
plt.plot(x,y1)
plt.grid(True)


```

```python
y = np.sin(x+(math.pi/4))
plt.plot(x,y)
plt.grid(True)
```

```python
y = np.e**(-0.1*x)* np.sin(3*x)
plt.plot(x,y)
```

```python
y = np.sin(x)*np.cos(2*x)
plt.plot(x,y)
```
